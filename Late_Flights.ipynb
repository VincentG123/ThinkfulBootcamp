For the interest of saving time, I am going to take just 2000 rows and work from there. I'll first shuffle the rows to get a randomized sample...

```python
late1_dummies = late1_dummies.sample(frac=1).reset_index(drop=True)
```

```python
late1_sample = late1_dummies[:2000]
```

```python
late1_sample.shape
```

```python
late1_sample = late1_sample.replace([np.inf, -np.inf], np.nan).dropna(axis=0)
```

```python
late1_sample.head()
```

```python
late1_sample.shape
```

```python
Y_train = late1_sample['ArrDelay'].values.reshape(-1, 1)
X_train = late1_sample.loc[:, ~(late1_sample.columns).isin(['ArrDelay'])]
```

```python
from sklearn import linear_model
from sklearn import preprocessing
```

```python
# Small number of parameters.
lass = linear_model.Lasso(alpha=.35)
lassfit = lass.fit(X_train, Y_train)
print('R for the model with few features:')
print(lass.score(X_train, Y_train))

print(len(X_train.columns))
origparams = np.append(lassfit.coef_, lassfit.intercept_)

print('\nParameter estimates for the model with few features:')
print(origparams)

for num, col in enumerate(X_train.columns):
    print(X_train.columns[num], ' = ', lassfit.coef_[num] )
```

At 99%, I'm certain this is WAY TOO over fitted... And I think it's because have the actually arrival time makes it too easy to figure out... So, I'm going to drop the arrival time and see what happens...

```python
Y_train = late1_sample['ArrDelay'].values.reshape(-1, 1)
X_train = late1_sample.loc[:, ~(late1_sample.columns).isin(['ArrDelay', 'ArrTime'])]

# Small number of parameters.
lass = linear_model.Lasso(alpha=.35)
lassfit = lass.fit(X_train, Y_train)
print('R for the model with few features:')
print(lass.score(X_train, Y_train))

print(len(X_train.columns))
origparams = np.append(lassfit.coef_, lassfit.intercept_)

print('\nParameter estimates for the model with few features:')
print(origparams)

for num, col in enumerate(X_train.columns):
    print(X_train.columns[num], ' = ', lassfit.coef_[num] )
```

Not much of a change... I'm going to drop all of the arrival and departure time information... And see what happens...

```python
Y_train = late1_sample['ArrDelay'].values.reshape(-1, 1)
X_train = late1_sample.loc[:, ~(late1_sample.columns).isin(['ArrDelay', 'ArrTime','DepTime','CRSDepTime','CRSArrTime'])]

# Small number of parameters.
lass = linear_model.Lasso(alpha=.35)
lassfit = lass.fit(X_train, Y_train)
print('R for the model with few features:')
print(lass.score(X_train, Y_train))

print(len(X_train.columns))
origparams = np.append(lassfit.coef_, lassfit.intercept_)

print('\nParameter estimates for the model with few features:')
print(origparams)

for num, col in enumerate(X_train.columns):
    print(X_train.columns[num], ' = ', lassfit.coef_[num] )
```

hmmmm... still 99%. The biggest indicators so far seem to be AirTime, DepDelay, TaxiIn, and TaxiOut. These actually seem the most obvious since if you leave 20 minutes late, you'd probably arrive 20 minutes late. Since it's so obvious, I'm going to drop that out too and see what happens... 

```python
Y_train = late1_sample['ArrDelay'].values.reshape(-1, 1)
X_train = late1_sample.loc[:, ~(late1_sample.columns).isin(['ArrDelay', 'ArrTime','DepTime','CRSDepTime','CRSArrTime','DepDelay'])]

# Small number of parameters.
lass = linear_model.Lasso(alpha=.35)
lassfit = lass.fit(X_train, Y_train)
print('R for the model with few features:')
print(lass.score(X_train, Y_train))

print(len(X_train.columns))
origparams = np.append(lassfit.coef_, lassfit.intercept_)

print('\nParameter estimates for the model with few features:')
print(origparams)

for num, col in enumerate(X_train.columns):
    print(X_train.columns[num], ' = ', lassfit.coef_[num] )
```

After eliminating some of the more obvious ones, CarrierDelay, WeatherDelay, NASDelay, LateAircraftDelar are some pretty strong indicators with a 1:1 ration in minutes. So for every 1 mintute delay caused by weather there is a 1 minute delay in arrival time.
