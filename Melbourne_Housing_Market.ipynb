```python
from sklearn import linear_model
from sklearn import preprocessing
```

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, shuffle=True)

for train_index, test_index in kf.split(housing1):
    data = housing1.iloc[train_index, housing1.columns != 'Price']
    target = housing1.iloc[train_index]['Price']
    test_data = housing1.iloc[test_index, housing1.columns != 'Price']
    test_target = housing1.iloc[test_index]['Price']
    
    housing1_columns = housing1.iloc[train_index, housing1.columns != 'Price'].columns
    
    lassBig = linear_model.Lasso(alpha=.001)
    lassBig.fit(data, target)
    print('\nRÂ² for the model with many features:')
    print(lassBig.score(data, target))
    print(lassBig.score(test_data, test_target))
#     origparams = np.append(lassBig.coef_, lassBig.intercept_)
#     print('\nParameter estimates for the model with many features:')
#     print(origparams)
    
    for int in range(len(lassBig.coef_)):
        print(housing1_columns[int], ' = ', lassBig.coef_[int])
    
```
