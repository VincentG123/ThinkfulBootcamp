With all this in mind, I was able to discover some interesting information.

Here is a violin plot of the domestic gross net of all the movies:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Imported the data
movie_metadata = pd.read_csv('movie_metadata.csv')

#This is how I got rid of the foreign currency. I removed any budget of $300,000,000.00
movie_metadata = movie_metadata.loc[movie_metadata.budget < 300000000] 

#Created a new "net" column from the gross - budget
result = movie_metadata.gross - movie_metadata.budget
movie_metadata['net'] = result
movie_metadata_net_array = np.asarray(movie_metadata.net.dropna())

#Used the net to create a violin plot
plt.violinplot(movie_metadata_net_array)
plt.title('Net Earnings Violin Plot')
plt.ylabel('100 Millions')
plt.show()
```

With this plot, I sadly came to realize why Hollywood executives are so adverse to risk: the vast majority of movies simply don't make any money. 

They're afraid of losing on their investment. And to be honest, I can't blame them. I agonize over decisions that would only cost me a few hundred dollars. I couldn't imagine making choices that could mean 100s of millions of dollars in lost revenue.

For more anaysis of the data, I also plotted the net profit as both a histogram and a box plot as well as calculated the mean, standard deviation, and the median.

```python
import locale
locale.setlocale( locale.LC_ALL, '' )

plt.figure(figsize=(15, 6))

#Histogram subplot
plt.subplot(1, 2, 1)
plt.hist(movie_metadata_net_array, color="Orange", bins=50)
plt.ylabel('Number of Movies')
plt.xlabel('100 Millions')
plt.title('Profit Histogram')
plt.axvline(movie_metadata_net_array.mean(), linestyle='solid', linewidth=2)
plt.axvline(movie_metadata_net_array.mean() + movie_metadata_net_array.std(ddof=1), 
            linestyle='dashed', linewidth=2)
plt.axvline(movie_metadata_net_array.mean()- movie_metadata_net_array.std(ddof=1), 
            linestyle='dashed', linewidth=2) 

#Boxplot subplot
plt.subplot(1, 2, 2)
plt.boxplot(movie_metadata_net_array,0, 'o', 0)
plt.axvline(x= np.median(movie_metadata_net_array),linewidth=1, color = 'Orange')
plt.xlabel('100 Millions')
plt.title('Profit Boxplot')
plt.show()

#Print mean, standard deviation, and median
print("Net Profit Mean = ", locale.currency(movie_metadata_net_array.mean(),grouping=True)  )
print('Net Profit Standard Deviation = ', locale.currency(movie_metadata_net_array.std(ddof=1),grouping=True) )
print('Net Profit Median = ', locale.currency(np.median(movie_metadata_net_array),grouping=True) )
```

Looking at the graphs, I was startled by how many movies just barely broke even hovering near and above zero dollars. Of course, this is all relatively speaking since the mean profit amount of \$13,385,285.79 and the median of \$999,737.50 are both way more than what most people will ever see in their lifetime. ### Cast a big Hollywood star

I know that the concept of what makes a "big Hollywood star" could be debatable, but for now, let's just go with somebody who is a household name, has been the lead in several big budget movies, and is known all over the world.

Below is a side by side comparison of box plots that compare the median net earnings of big Hollywood movie stars.

```python
pd.options.mode.chained_assignment = None

font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }

#Created a function that would make plot with mulitple boxplots
def star_box(star_list):
    star_nets = [movie_metadata_net_array]
    star_medians = 'Medians:\nAll Movies ' + \
    locale.currency(np.median(movie_metadata_net_array), grouping=True)
    
    #Loop through the star_list to create the dataframes for the boxplots and the text
    #for the legend.
    for star in star_list:
        star_df = movie_metadata.loc[(movie_metadata.actor_1_name == star) | 
                                     (movie_metadata.actor_2_name == star) | 
                                     (movie_metadata.actor_3_name == star)]
        star_df['net'] = star_df.gross - star_df.budget
        star_nets.append(np.asarray(star_df.net.dropna()))
        star_medians += '\n' + star + ' '+ \
        locale.currency(np.median(np.asarray(star_df.net.dropna())), grouping=True)
    
    #Plotted the boxplot
    plt.figure(figsize=(12,12))
    plt.boxplot(star_nets)
    
    #Added the textbox
    plt.text(2, 300000000, star_medians,bbox={'facecolor':'red', 'alpha':0.5, 'pad':10})
    
    #Renamed x-axis tickmarks
    star_list = ["All Movies"] + star_list
    plt.xticks(range(1,len(star_list)+1),star_list, rotation=45)
    
    #Label the y-axis
    plt.ylabel('100 Millions')
    
    #Added a line that is the median line for "All Movies"
    plt.axhline(y= np.median(movie_metadata_net_array),linewidth=1, color = 'Green')
    
    #Add title
    plt.title('Median Net Earnings of Movies per Hollywood Star', fontdict=font)

    #Label the different outlier movies
    plt.text(9.4, 290000000, 'Forest Gump')
    plt.text(9.4, -190000000, 'Polar Express')
    plt.text(6.8, 470000000, 'Titanic')
    plt.text(0.8, 540000000, 'Avatar')
    plt.text(0.8, -210000000, 'John Carter')
    plt.text(2.7, 120000000, 'Magic Mike')
    plt.text(1.7, -140000000, 'Alexander')
    plt.text(4.5, 180000000, 'Pretty Woman')
    
    plt.show()
    
#Called the function with a list of stars
star_box(['Angelina Jolie Pitt', 'Matthew McConaughey', 'Meryl Streep',  'Julia Roberts',
            'Will Smith', 'Leonardo DiCaprio', "Emma Watson", 'Tom Cruise', 'Tom Hanks'])
```

As you can see, all of these known Hollywood actors are above the "All Movies" median except for Angelina Jolie Pitt. Also, all of these actors have movies that have made a net profit of over 100 million dollars except for the world renouned Meryl Streep and (once again) Angelina Jolie Pitt. Both, however, came very close. Jolie Pitt with her turn as a sexy fish named Lola in the animated "Shark Tale" raked in a hefty \$85,762,022.00, while Streep's leading role in the musical "Mamma Mia" brought in an even heftier \$91,704,210.00.

Overall, the big stars do very well. Will Smith, for example, has broken the \$200,000,000.00 million mark with "Independance Day" while Tom Hanks came very close to \$300,000,000 with "Forest Gump". Leonardo DiCaprio, however, is the box office champ with "Titanic" just shy of the \$500,000,000.00 dollar threshold.

Unfortunately even with all of this success, there are no guarantees. It's still a risky venture where established stars like Tom Hanks can lead a movie to a net loss of \$164,334,574.00. Movie executives need to remember that for every "Forest Gump" there could also exist a "Polar Express".

### Stick to a Certain Genre

I've assumed for a while now that in order to make the hundreds of millions of dollars that Hollywood craves, one would have to stick with big, action packed sci-fi/fantasy blockbusters.

And I was partially right.

Here is a grid containing histograms of movies that were grouped according to genre in order of mean net earnings:

```python
import matplotlib.lines as mlines
import matplotlib.pyplot as plt

#Function to get the nets of different genres
def genre_histogram(genre):
    temp_df = movie_metadata.loc[movie_metadata.genres.str.contains(genre)]
    temp_df_net = temp_df.gross - temp_df.budget
    temp_df['net'] = temp_df_net
    return temp_df.net.dropna()

blue_line = mlines.Line2D([], [], color='blue', label='All Movies Mean Earnings')
green_line = mlines.Line2D([], [], color='green', markersize=15, label='Genre Mean Earnings')

handles = [blue_line, green_line]
labels = [l.get_label() for l in handles] 
# fig.legend(handles=handles, labels=labels)  

text2 = 'Mean Earnings:'
#Creating all of the different subplots
fig = plt.figure(figsize= (15,12))
fig.suptitle('Mean Net Earnings of Movies per Genre', fontdict=font)
fig.text(0.5, 0.02, '100 Millions', ha='center')
fig.legend(handles = handles, labels = labels)

# fig.legend((l3, l4), ('Line 3', 'Line 4'), 'upper right')

ax10 = fig.add_subplot(3,4,1)
x = genre_histogram("Documentary")
ax10.hist(x, color="Orange", bins=50)
ax10.set_xlim(-75000000, 100000000)
ax10.set_title('Documentary')
ax10.axvline(movie_metadata.net.mean(), color='b', linestyle='solid', linewidth=.5)
ax10.axvline(x.mean(), color='Green', linestyle='solid', linewidth=.5)
text2 +='\nDocumentary = ' + locale.currency(x.mean(), grouping=True)

ax3 = fig.add_subplot(3,4,2)
x = genre_histogram("Drama")
ax3.hist(x, color="Orange", bins=50)
ax3.set_xlim(-75000000, 100000000)
ax3.set_title('Drama')
ax3.axvline(movie_metadata.net.mean(), color='b', linestyle='solid', linewidth=.5)
ax3.axvline(x.mean(), color='Green', linestyle='solid', linewidth=.5)
text2 +='\nDrama = ' + locale.currency(x.mean(), grouping=True)

ax9 = fig.add_subplot(3,4,3)
x = genre_histogram("Action")
ax9.hist(x, color="Orange", bins=50)
ax9.set_xlim(-75000000, 100000000)
ax9.set_title('Action')
ax9.axvline(movie_metadata.net.mean(), color='b', linestyle='solid', linewidth=.5)
ax9.axvline(x.mean(), color='Green', linestyle='solid', linewidth=.5)
text2 +='\nAction = ' + locale.currency(x.mean(), grouping=True)

ax11 = fig.add_subplot(3,4,4)
x = genre_histogram("Horror")
ax11.hist(x, color="Orange", bins=50)
ax11.set_xlim(-75000000, 100000000)
ax11.set_title('Horror')
ax11.axvline(movie_metadata.net.mean(), color='b', linestyle='solid', linewidth=.5)
ax11.axvline(x.mean(), color='Green', linestyle='solid', linewidth=.5)
text2 +='\nHorror = ' + locale.currency(x.mean(), grouping=True)

ax1 = fig.add_subplot(3,4,5)
ax1.set_facecolor('red')
ax1.hist(movie_metadata_net_array, color="Orange", bins=50)
ax1.set_xlim(-75000000, 100000000)
ax1.set_title('All Movies')
ax1.axvline(movie_metadata.net.mean(), color='b', linestyle='solid', linewidth=.5)
text2 +='\nAll Movies = ' + locale.currency(movie_metadata.net.mean(), grouping=True)

ax6 = fig.add_subplot(3,4,6)
x = genre_histogram("Romance")
ax6.hist(x, color="Orange", bins=50)
ax6.set_xlim(-75000000, 100000000)
ax6.set_title('Romance')
ax6.axvline(movie_metadata.net.mean(), color='b', linestyle='solid', linewidth=.5)
ax6.axvline(x.mean(), color='Green', linestyle='solid', linewidth=.5)
text2 +='\nRomance = ' + locale.currency(x.mean(), grouping=True)

ax4 = fig.add_subplot(3,4,7)
x = genre_histogram("Comedy")
ax4.hist(x, color="Orange", bins=50)
ax4.set_xlim(-75000000, 100000000)
ax4.set_title('Comedy')
ax4.axvline(movie_metadata.net.mean(), color='b', linestyle='solid', linewidth=.5)
ax4.axvline(x.mean(), color='Green', linestyle='solid', linewidth=.5)
text2 +='\nComedy = ' + locale.currency(x.mean(), grouping=True)

ax2 = fig.add_subplot(3,4,8)
x = genre_histogram("Sci-Fi")
scifi_mean = x.mean()
scifi_length = len(x)
ax2.hist(x, color="Orange", bins=50)
ax2.set_xlim(-75000000, 100000000)
ax2.set_title('Sci-Fi')
ax2.axvline(movie_metadata.net.mean(), color='b', linestyle='solid', linewidth=.5)
ax2.axvline(x.mean(), color='Green', linestyle='solid', linewidth=.5)
text2 +='\nSci-Fi = ' + locale.currency(x.mean(), grouping=True)

ax7 = fig.add_subplot(3,4,9)
x = genre_histogram("Musical")
musical_mean = x.mean()
musical_length = len(x)
ax7.hist(x, color="Orange", bins=50)
ax7.set_xlim(-75000000, 100000000)
ax7.set_title('Musical')
ax7.axvline(movie_metadata.net.mean(), color='b', linestyle='solid', linewidth=.5)
ax7.axvline(x.mean(), color='Green', linestyle='solid', linewidth=.5)
text2 +='\nMusical = ' + locale.currency(x.mean(), grouping=True)

ax5 = fig.add_subplot(3,4,10)
x = genre_histogram("Fantasy")
fantasy_mean = x.mean()
fantasy_length = len(x)
fantasy_std = x.std(ddof=1)
ax5.hist(x, color="Orange", bins=50)
ax5.set_xlim(-75000000, 100000000)
ax5.set_title('Fantasy')
ax5.axvline(movie_metadata.net.mean(), color='b', linestyle='solid', linewidth=.5)
ax5.axvline(x.mean(), color='Green', linestyle='solid', linewidth=.5)
text2 +='\nFantasy = ' + locale.currency(x.mean(), grouping=True)

ax8 = fig.add_subplot(3,4,11)
x = genre_histogram("Family")
family_mean = x.mean()
family_length = len(x)
family_std = x.std(ddof=1)
ax8.hist(x, color="Orange", bins=50)
family_length = len(x)
ax8.set_xlim(-75000000, 100000000)
ax8.set_title('Family')
ax8.axvline(movie_metadata.net.mean(), color='b', linestyle='solid', linewidth=.5)
ax8.axvline(x.mean(), color='Green', linestyle='solid', linewidth=.5)
text2 +='\nFamily = ' + locale.currency(x.mean(), grouping=True)

ax12 = fig.add_subplot(3,4,12)
x = genre_histogram("Animation")
animation_length = len(x)
animation_mean = x.mean()
animation_std = x.std(ddof=1)
ax12.hist(x, color="Orange", bins=50)
ax12.set_xlim(-75000000, 100000000)
ax12.set_title('Animation')
ax12.axvline(movie_metadata.net.mean(), color='b', linestyle='solid', linewidth=.5)
ax12.axvline(x.mean(), color='Green', linestyle='solid', linewidth=.5)
text2 +='\nAnimation = ' + locale.currency(x.mean(), grouping=True)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])

plt.show()
```

Sci-fi and Fantasy definitely rank above the "All Movies" mean earnings of \$13,385,285.79. I was surprised, however, that they were both outclassed by Family and Animation. I was also surprised that Musicals made a strong showing sneeking in between Sci-Fi and Fantasy. But once it all sunk in, it made complete sense.

I had forgotten about the slew of Disney movies that have been a mainstay for decades. Ever since Snow White first wished upon a star, Disney princesses have been spinning box office gold. Being a mixture of Musical, Animation, and Fantasy, they enjoy an advantage over other genres. And the most current break away hit "Frozen" proves that point by  earning a net profit of \$250,736,600.00. With numbers like that, it looks to be that animated movie musicals are the way to go!

But, let's not forget that live action Fantasy and Sci-fi movies boast some of the highest earning movies of all time with the likes of "Avatar", "Jurassic World", "Star Wars", The Lord of the Rings Trilogy, and the Marvel Cinematic Universe. 

Taking that into consideration and because there are fewer animation movies made as compared to fantasy, I decided to compute the standard errors to see if the mean earnings still held up:

```python
print("Fantasy Mean Earnings = " + locale.currency(fantasy_mean , grouping=True))
print("Fantasy Standard Error = +/- " + locale.currency(fantasy_std/fantasy_length**0.5 , grouping=True))
print("Animation Mean Earnings = " + locale.currency(animation_mean , grouping=True))
print("Animation Standard Error = +/- " + locale.currency(animation_std/animation_length**0.5 , grouping=True))
overlap = (fantasy_mean + fantasy_std/fantasy_length**0.5) - (animation_mean - animation_std/animation_length**0.5)
print("Possible overlap = " + locale.currency(overlap , grouping=True))
```

There is some overlap with the standard error. I think, though, it is safe to conclude that animated movies are a slightly better bet. 

### Actually Make a Good Movie

I know Hollywood isn't known for putting art above profit, but I was curious to see if the quality of a movie had any corralation to it's net earnings. 

In this scatter plot, I am using a movie's IMDB score as an indicator of how good the movie is and graphing it against the net earnings.

```python
#Creating scatter plot of net vs imdb scores
plt.figure(figsize= (15,15))
plt.scatter(movie_metadata.imdb_score, movie_metadata.net)
plt.title('Net Earning vs IMDB Scores', fontdict=font)

#Added lines to show section of movies that made over $100,000,000 and scored higher than a 4
plt.axhline(100000000, color='orange', linestyle='solid')
plt.axvline(4, color='orange', linestyle='solid')

#Add labels
plt.ylabel('100 Million')
plt.xlabel('IMDB Scores')

plt.grid()
plt.show()
```

One of the take aways I gleened from this plot is that no movie that grossed over \$100,000,000.00 scored lower that a 4 on their IMDB score. So basically, nothing truly horid will ever be a box office darling.

That being said, box office winners who scored between 4 to 6 were movies based on exisiting properties such as "Twilight", "Transformers", and "Fifty Shades of Grey". These properties already had massive fanbases who were clamoring to see their favorite characters brought to life on the big screen.  

This trend continues in the 6 to 7 range as well, but this grouping also includes breakout hits that stand alone without the help of being an already known entity. These movies include unexpected hits like "Bridesmaids", "The Blair Witch Project", and "Fatal Attraction". Movies attached to franchises, though, can really score big with the likes of "Star Wars: Episode 1- The Phantom Menace" earning above the \$300,000,000.00 mark.

To be in the above \$400,000,000.00 club though, you must score at least a 7.0. This elite group of 7 movies contain 3 franchise films with "Jurrasic World", "The Avengers", and "The Avengers- Age of Ultron", 2 stand alones with "Titanic" and "E.T. the Extra-Terrestrial", and 2 monster hits that have already or will soon spawn massive franchises with "Star Wars: Episode IV- A New Hope" and "Avatar".

Fortunately for Hollywood, you don't necessarily have to go 8 or higher. Box office champ "Avatar" sits comfortably on top with a respectable 7.9 rating. (Once again, I really didn't like this movie). And of the 4 movies that have scored a 9.0 or higher, only 2 have made over \$100,000,000.00 with "The Dark Knight" and "The Godfather".
